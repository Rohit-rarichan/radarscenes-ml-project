```python
import matplotlib.pyplot as plt
import numpy as np

# --- Replace these arrays with your real loss values ---
epochs = np.arange(1, 21)
train_loss = np.linspace(0.2, 1.0, 20) + np.random.normal(0, 0.03, 20)
val_loss = 0.25 + 0.1*np.sin(np.linspace(0, 3*np.pi, 20)) + np.linspace(0.0, 0.4, 20)

# --- Style settings to mimic your slide ---
plt.style.use("dark_background")
fig, ax = plt.subplots(figsize=(10,6))

ax.plot(epochs, train_loss, color="#4A8BFF", linewidth=3, label="Training Loss")
ax.plot(epochs, val_loss, color="#FFC94A", linewidth=3, label="Validation Loss")

# Smooth curve look (optional)
# If your values are noisy, uncomment:
# from scipy.ndimage import gaussian_filter1d
# train_loss = gaussian_filter1d(train_loss, sigma=2)
# val_loss = gaussian_filter1d(val_loss, sigma=2)

# Axes styling
ax.set_facecolor("#1A1F2B")       # Dark navy-like background
fig.patch.set_facecolor("#1A1F2B")

ax.set_xlabel("Epochs", fontsize=14, color="white")
ax.set_ylabel("Loss", fontsize=14, color="white")
ax.tick_params(colors="white", labelsize=12)

# Grid (subtle)
ax.grid(color="white", alpha=0.1, linestyle="--")

# Legend
ax.legend(fontsize=14)

plt.tight_layout()
plt.show()
```
